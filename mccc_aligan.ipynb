```python
import obspy
```

```python
strm = obspy.core.Stream()
datadir = 'synsac/'
with open(datadir+'tdel.dat') as filein:
    lines = filein.readlines()
    for iline in lines:
        traceid,shift,cc = iline.split()
        trace = obspy.read(datadir+traceid)[0]
        trace.stats.distance = int(traceid.split('.')[0])*1000
        strm.append(trace)
        
```

```python
strm.plot(type='section',orientation='horizontal');
```

```python
strm = obspy.core.Stream()
import numpy as np
stack = np.zeros(200,)
with open(datadir+'tdel.dat') as filein:
    lines = filein.readlines()
    for iline in lines:
        traceid,shift,cc = iline.split()
        trace = obspy.read(datadir+traceid)[0]
        trace.data = np.roll(trace.data,int(float(shift)*10))
        trace.stats.distance = int(traceid.split('.')[0])*1000
        strm.append(trace)
        stack += trace.data
```

```python
strm.plot(type='section',orientation='horizontal');
```

```python
from matplotlib import pyplot as plt
plt.plot(stack)
```
